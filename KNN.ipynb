```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#for KNN
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df_merge = pd.read_csv('Merged0430.csv')
df_merge.info()
```

```python
df_merge.shape
```

```python
#check null values
df_merge.isnull().sum()
```

```python
pd.set_option('display.max_columns', None)
df_merge.head()
```

```python
df = df_merge.drop(columns = ['Accident_Index', 'Accident_Severity3Category', 'Accident_Severity2Category', 'Date', 'Month','Year', 'Day_of_Week', 'Time', 'Hour', 'Weather_Conditions', 'Precipitation',
                               'RoadSurfaceNumeric', 'Vehicle_Reference'], axis = 1)
df.head()
```

```python
df_encoded = pd.get_dummies(df)
df_encoded = df_encoded.astype(int)
df_encoded.head()
```

```python
df.shape
```

```python
df_encoded.shape
```

```python
#separate independent and dependent variables
X = df_encoded.drop('Accident_Severity', axis =1)
y = df_encoded['Accident_Severity']
```

```python
#normalise independent variables
min_max_scaler = MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X)
X_scaled
```

```python
#split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
# Predict using the test set
y_pred = knn.predict(X_test)
```

```python
# Print model accuracy score
accuracy_score_1 = accuracy_score(y_test, y_pred)
print(f"Accuracy ={round((accuracy_score_1 * 100), 4)}%")
```

```python
# Set up array of possible k values
possible_k = np.arange(50,500,100)
possible_k
```

```python
# Calculate accuracy for each k value
ac_scores=[]
for k in possible_k:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    scores=round((accuracy_score(y_test, y_pred) * 100), 6)
    ac_scores.append(scores)
print ("Accuracy Scores :", ac_scores)
```
